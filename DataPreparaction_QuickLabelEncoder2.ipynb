```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv('loan_data_cleaned.csv')
dataset.head(10)
```

```python
dataset.select_dtypes(object)
```

```python
for_en_data = dataset[['Gender', 'Married', 'Dependents', 'Education','Self_Employed','Property_Area', 'Loan_Status']]
```

```python
for xe in for_en_data:
    print('****************************')
    print(for_en_data[xe].value_counts())
```

```python
le = LabelEncoder()
```

```python
EnCoded_data = pd.DataFrame()
for x in for_en_data:
    EnCoded_data['en_'+x] = le.fit_transform(for_en_data[x])

EnCoded_data.head(15)
```
